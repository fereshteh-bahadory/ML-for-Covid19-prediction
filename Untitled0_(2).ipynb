```python
data2=pd.read_csv(io.BytesIO(uploaded2['country_vaccinations_by_manufacturer.csv']))
data2.head()
```

```python
data2.info()
```

```python
data2.replace({'vaccine':{'Johnson&Johnson':1,'Moderna':2,'Oxford/AstraZeneca':3,'Pfizer/BioNTech':4,
                          'Sinovac':5,'CanSino':6,'Sputnik V':7,'Sinopharm/Beijing':8}},inplace=True)
data2.info()
print("Johnson&Johnson:1,Moderna:2,Oxford/AstraZeneca:3,Pfizer/BioNTech:4,Sinovac:5,CanSino:6,SputnikV:7,Sinopharm/Beijing:8")
```

```python
data2.set_index('date',inplace=True)
data2.sort_values(by="date").reset_index(drop=True)
```

```python
fig, ax = plt.subplots(figsize=[7,5])
sb.distplot(data2['total_vaccinations'],color='g', ax=ax)
ax.set_title('Total Vaccinations')
plt.show()
```

```python
data2['pre6_total_vaccinations']=data2['total_vaccinations'].shift(+6)
data2['pre8_total_vaccinations']=data2['total_vaccinations'].shift(+8)
data2['post_total_vaccinations']=data2['total_vaccinations'].shift(-1)
data2.fillna(0,inplace=True)
data2.head(10)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn import metrics
reg=LinearRegression()
```

```python
from sklearn.preprocessing import PolynomialFeatures
pl=PolynomialFeatures(degree=2)
```

```python
from sklearn.ensemble import RandomForestRegressor
rf=RandomForestRegressor(n_estimators=150,random_state=24)
```

```python
X1,X2,X3,y=data2['vaccine'],data2['pre8_total_vaccinations'],data2['pre6_total_vaccinations'],data2['post_total_vaccinations']
X1,X2,X3,y=np.array(X1),np.array(X2),np.array(X3),np.array(y)
X1,X2,X3,y=X1.reshape(-1,1),X2.reshape(-1,1),X3.reshape(-1,1),y.reshape(-1,1)
X=np.concatenate((X2,X3),axis=1)
X_train,X_test,y_train,y_test=X[:-50],X[-50:],y[:-50],y[-50:]
rf.fit(X_train,y_train)
```

```python
y_pred=rf.predict(X_test)
plt.rcParams["figure.figsize"]=(15,8)
plt.plot(y_pred,label="Prediction Values")
plt.plot(y_test,label="Actual Values")
plt.show()
```

```python
reg.fit(X_train,y_train)
y_pred=reg.predict(X_test)
plt.rcParams["figure.figsize"]=(15,8)
plt.plot(y_pred,label="Prediction Values")
plt.plot(y_test,label="Actual Values")
plt.show()
```
